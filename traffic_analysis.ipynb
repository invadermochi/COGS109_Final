```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

# Exploratory Analysis

```python
traffic_data = pd.read_csv("Metro_Interstate_Traffic_Volume.csv")

# Keeping only the relevant columns to our research question
data = traffic_data.drop(traffic_data.columns.difference(["temp", "clouds_all", "traffic_volume"]), axis=1)
data.head()
```

```python
# Looking at our variables and variances
f, axs = plt.subplots(1, 2, figsize=(10,10))

axs[0].hist(data["temp"])
axs[1].hist(data["clouds_all"])

axs[0].set_xlabel("Temperature in Kelvin")
axs[1].set_xlabel("Percentage of Cloud Cover")

plt.show()

print("The variance for temperature is: {}".format(data["temp"].var()))
print("The variance for cloud cover is: {}".format(data["clouds_all"].var()))
data.describe()

# Looking at this, do we need to scale the data guys?
```
